### Step 1.3: Create a Cleaning Function

Given that I will perform the same cleaning steps on the customer demographics data, I will write a function that executes the main feature selection, encoding, and re-engineering steps performed above. Therefore, I can just run this function on that DataFrame to get the trimmed dataset in a single step.

```python
def clean_data(df, feat_info):
    """
    Perform feature trimming, re-encoding, and engineering for demographics data
    
    INPUT: 
        df - demographics datagrame 
        feat_info - feature summary dataframe
           
    OUTPUT: 
        Trimmed and cleaned demographics datagrame
    """
    # Put in code here to execute all main cleaning steps:
    
    # Check if column variables in df are the same as attributes in feat_info
    result = list(df.columns) == list(feat_info['attribute'])
    
    if result:       
        # Convert missing value codes into null
        print('Converting missing code......')
        feat_info['missing_or_unknown'] = feat_info['missing_or_unknown'].str.strip('[]').str.split(',')
        
        for i in range(feat_info.shape[0]):
            lookup = ['X', 'XX', '']
            if any(e in lookup for e in feat_info['missing_or_unknown'][i]):
                pass
            else:
                feat_info['missing_or_unknown'][i] = [float(x) for x in feat_info['missing_or_unknown'][i]]
    
        for i, col in enumerate(list(df.columns)):
            df[col] = df[col].replace(feat_info['missing_or_unknown'][i], np.nan)

        # Remove outliers that have more than 200,000 missing values in each column
        print('Removing outliers......')        
        feat_info['missing'] = df.isnull().sum().tolist()
        outliers = list(feat_info[feat_info['missing'] > 200000]['attribute'])
        df = df.drop(outliers, axis=1)
        feat_info = feat_info[~feat_info['attribute'].isin(outliers)]

        # Remove outliers that have more than 20 missing values in each row
        df['missing'] = df.isnull().sum(axis=1).tolist()
        df = df[df['missing'] <= 20]      
            
        # Fill null with mode
        print('Imputing data......')
        fill_mode = lambda col: col.fillna(col.mode()[0])
        df = df.apply(fill_mode)
        
        # Assess and re-encode categorical variables 
        print('Re-encoding categorical variables......')        
        # Re-encode binary categorical variables
        df['OST_WEST_KZ'] = df['OST_WEST_KZ'].replace({'W':0, 'O':1})
                
        # Re-encode multi-level categorical variables    
        df = df.drop('CAMEO_DEU_2015', axis=1)
        
        CAMEO_DEUG_2015 = pd.get_dummies(df['CAMEO_DEUG_2015'], prefix='CAMEO_DEUG_2015')
        df = pd.concat([df, CAMEO_DEUG_2015], axis=1)
        df = df.drop('CAMEO_DEUG_2015', axis=1)
        
        # Engineer Mixed-Type Features
        # PRAEGENDE_JUGENDJAHRE
        print('Engineering variable "PRAEGENDE_JUGENDJAHRE"......')        
        to_replace = {1: [1, 1],
                      2: [1, 2],
                      3: [2, 1],
                      4: [2, 2], 
                      5: [3, 1], 
                      6: [3, 2], 
                      7: [3, 2], 
                      8: [4, 1],
                      9: [4, 2], 
                      10: [5, 1],
                      11: [5, 2], 
                      12: [5, 1], 
                      13: [5, 2], 
                      14: [6, 1], 
                      15: [6, 2]}
        df['PRAEGENDE_JUGENDJAHRE'] = df['PRAEGENDE_JUGENDJAHRE'].map(to_replace)
        df['PRAEGENDE_JUGENDJAHRE_GEN'] = df['PRAEGENDE_JUGENDJAHRE'].apply(lambda x: x[0])
        df['PRAEGENDE_JUGENDJAHRE_MOVE'] = df['PRAEGENDE_JUGENDJAHRE'].apply(lambda x: x[1])
        df = df.drop('PRAEGENDE_JUGENDJAHRE', axis=1)
        
        # CAMEO_INTL_2015
        print('Engineering variable "CAMEO_INTL_2015"......')        
        df['CAMEO_INTL_2015_WEALTH'] = df['CAMEO_INTL_2015'].apply(lambda x: x[0])
        df['CAMEO_INTL_2015_LIFE'] = df['CAMEO_INTL_2015'].apply(lambda x: x[1])
        df = df.drop('CAMEO_INTL_2015', axis=1)
        
        # Drop other columns with mixted types of data 
        print('Dropping other mixed variables......')       
        df = df.drop(['LP_LEBENSPHASE_FEIN', 'LP_LEBENSPHASE_GROB', 'WOHNLAGE', 'PLZ8_BAUMAX'], axis=1)
        
        # Drop variable missing
        df = df.drop('missing', axis=1)
        
    else:
        print('The column variables of the input dataframe are not consistent with the attributes of feat_info.')
        
    return df 
```

```python
# Define a function to visualize the variance explained by each principal component
def scree_plot(pca):
    """
    Creates a scree plot associated with the principal components.
    """
    n_comps = len(pca.explained_variance_ratio_)
    index = np.arange(n_comps)
    values = pca.explained_variance_ratio_
 
    plt.figure(figsize=(10, 6))
    ax = plt.subplot(111)
    cum_values = np.cumsum(values)
    ax.bar(index, values)
    ax.plot(index, cum_values)
    for i in range(5):
        ax.annotate(r"%s%%" % ((str(values[i]*100)[:4])), (index[i]+0.2, values[i]), va="bottom", ha="center", fontsize=12)
 
    ax.xaxis.set_tick_params(width=0)
    ax.yaxis.set_tick_params(width=2, length=12)
 
    ax.set_xlabel("Principal Component")
    ax.set_ylabel("Variance Explained (%)")
    plt.title('Explained Variance Per Principal Component')
```

### Step 2.3: Interpret Principal Components

With the transformed principal components, I will check out the weight of each variable on the first few components to see if they can be interpreted in some fashion.

Each principal component is a unit vector that points in the direction of highest variance (after accounting for the variance captured by earlier principal components). The further a weight is from zero, the more the principal component is in the direction of the corresponding feature. If two features have large weights of the same sign (both positive or both negative), then increases in one tend expect to be associated with increases in the other. To contrast, features with different signs can be expected to show a negative correlation: increases in one variable should result in decreases in the other.

- To investigate the features, I will map each weight to their corresponding feature name, then sort the features according to weight. The most interesting features for each principal component will then be those at the beginning and end of the sorted list. 
- I will write a function that prints the sorted list of feature weights, for the *i*-th principal component. This will help facilitate the investigation and interpretation of feature associations from the first three principal components. 

```python
# Create a dataframe to show each weight to their corresponding feature name
def pca_result(pca, df):
    """
    Create a dataframe df_comp to show weights to their corresponding features of each component/dimension.
    """
    dimensions = [f'Dimension {i}' for i in range(1, len(pca.components_) + 1)]

    df_comp = pd.DataFrame(np.round(pca.components_, 4))
    df_comp.columns = list(df.columns)
    df_comp.index = dimensions

    # Add variance ratio explained by each component/dimension
    variance_ratios = pca.explained_variance_ratio_.reshape(len(pca.components_), 1)
    df_variance = pd.DataFrame(np.round(variance_ratios, 4))
    df_variance.columns = ['Explained Variance']
    df_variance.index = dimensions

    df_comp = pd.concat([df_variance, df_comp], axis=1)
    
    return df_comp
```

```python
azdias_comp = pca_result(pca_20, azdias_scaled)
azdias_comp.head()
```

```python
# Define a function to show a sorted list of feature weights
def sort_features(dim, df):
    """
    Return the sorted list of feature weights for the number of dimension/principal component.
    """
    return df.iloc[dim-1,:].sort_values()
```

```python
# Investigate the change in within-cluster distance across number of clusters.
# HINT: Use matplotlib's plot function to visualize this relationship.

# Define a function to plot the result across number of clusters
def elbow_method(n_cluster, data):
    """
    Plot the kmeans scores of data across number of clusters.
    
    Input: n_cluster - number of clusters 
           data
    """
    centers = list(range(1, n_cluster+1))
    scores = []
    for center in centers:
        kmeans = KMeans(center)
        model = kmeans.fit(data)
        score = np.absolute(model.score(data))
        scores.append(score)
    
    plt.plot(centers, scores, linestyle='--', marker='o', color='b')
    plt.xlabel('K')
    plt.ylabel('Score')
    plt.title('Score vs. Number of Clusters');
```

```python
# Compare the proportion of data in each cluster for the customer data to the
# proportion of data in each cluster for the general population.

# Create a dataframe to store proportion of data points of each cluster for the general population 
# and the customer data
general_ls = []
customer_ls = []
clusters = range(10)
for cluster in clusters:
    general_prop = (azdias_clust == cluster).sum() / len(azdias_clust)
    general_ls.append(general_prop)
    customer_prop = (customers_clust == cluster).sum() / len(customers_clust)
    customer_ls.append(customer_prop)
    
cluster_dict = {'cluster': clusters,
               'general_prop': general_ls, 
               'customer_prop': customer_ls}

df_cluster = pd.DataFrame(cluster_dict)
df_cluster
```
